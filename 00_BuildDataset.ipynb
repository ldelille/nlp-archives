# Build the dataset

The goal of this notebook is to pave the way for data analysis by building the dataset. The news articles are contained in .json files inside subfolders. Here we collect the articles and organize them in a pandas dataframe, reusable as .csv file.

__Imports__

```python
import os
import glob
import json
import pandas as pd
from collections import OrderedDict
```

__Data collection__

```python
# Empty lists for collecting data
title=[]
text=[]
year=[]
url=[]

# open json files and extract data
for folder in os.listdir('./InputArticles/'):
    if folder[0] in '01': # exclude .DS_Store folder, check that folder name is a date
        print(folder)
        folderpath=os.path.join('./InputArticles/', folder)
        for filename in os.listdir(folderpath):
            filepath=os.path.join(folderpath, filename)
            print(filepath)
            #json_file = open(filepath)
            #j = json.loads(json_file)
            with open(filepath) as f:
                data = json.loads(f.read())
                title.append(data[0]['title'])
                text.append(data[0]['text'])
                url.append(data[0]['url'])
                year.append(int(folder)) # folder name = year
```

```python
# build dataframe
news_df=pd.DataFrame()
news_df['year']=year
news_df['title']=title
news_df['text']=text
news_df['url']=url
```

__Save the dataset__

```python
# save dataframe as .csv for future use
news_df.to_csv("articles.csv")

# save dataframe as Excel file for easy manual labeling
news_df.to_excel("articles_labeled.xlsx")
```
